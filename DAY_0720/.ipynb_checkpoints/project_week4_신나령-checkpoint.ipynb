## 1. 미국 금리 2019~2022 전처리

```python
import pandas as pd
import platform
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings(action='ignore')

amDF=pd.read_csv('미국금리_월.csv',encoding='euc_kr')
amDF.head()
```

```python
# 결측치 있는지 확인
amDF.isnull().sum()
```

```python
amDF.info()
```

```python
# 값만 따오고 싶어서 % 없애고 실수형으로 data 바꿔줌
amDF['실제']=amDF['실제'].apply(lambda x : x.replace('%','')).astype('float')
```

```python
dateDF=amDF['날짜'].apply(lambda x : x.split('-')[:2]) # 연, 월, 일 분리 ('-' 기준)
dateDF=pd.DataFrame(dateDF.tolist(),columns=('연','월')) # list 형태로 바꾼 걸 다시 DF로 생성
```

```python
# 연, 월, 일 분리한 DF에 원데이터 '실제' 컬럼 연결
amDF=pd.concat([dateDF,amDF['실제']],axis=1)
amDF.head()
```

```python
# 2019년 데이터
amDF_2019=amDF[amDF.연=='2019']
# 2020년 데이터
amDF_2020=amDF[amDF.연=='2020']
# 2021년 데이터
amDF_2021=amDF[amDF.연=='2021']
# 2022년 데이터
amDF_2022=amDF[amDF.연=='2022']
amDF_2019.head()
```

```python
# 금리 '연', '월' data 사이에 '-' 넣어서 저장해 주는 함수
def interest_rate(data):
    data['날짜']=data.apply(lambda x:str(x['연'])+'-'+x['월'],axis=1)
    data=data.iloc[:,2:]
    data.set_index('날짜',inplace=True)
    return data
```

```python
amDF_2019=interest_rate(amDF_2019)
amDF_2020=interest_rate(amDF_2020)
amDF_2021=interest_rate(amDF_2021)
amDF_2022=interest_rate(amDF_2022)
```

```python
amDF_total=pd.concat([amDF_2019,amDF_2020,amDF_2021,amDF_2022])
amDF_total.head()
```

```python
# 금리 ,코스피, 코스닥 그래프 그리는 함수
def draw_graph(data,name):    
    if(platform.system()=='Windows'):
        plt.rc('font',family='Malgun Gothic')
    else:
        plt.rc('font',family='AppleGothic')       
    plt.rcParams['axes.unicode_minus']=False
    plt.figure(figsize=(20,10))
    plt.title('2019~2022년 {}'.format(name),size=20)
    plt.plot(data,'orange',marker='o')
    plt.xticks(data.index,rotation=60,size=13)
    plt.xlabel('기간',size=15)
    if name=='미국 금리': 
        plt.ylabel('금리',size=15)
        plt.plot(data,'orange',marker='o',label=name+' 2019~2022')
        plt.annotate('<= 2020년 3월 코로나 대유행',xy=(15.1,0.25),rotation=20,fontsize=16)
        plt.annotate('',xy=(40,1.3),xytext=(37,0.4),xycoords='data',arrowprops=dict(arrowstyle='->',color='r',lw=2))
        plt.annotate('금리 인상 =>',xy=(36,0.7),rotation=70,fontsize=16)
    else:
        plt.ylabel('종가',size=15)
        plt.plot(data,'green',marker='o',label=name+' 2020년 3~5월')
        plt.title('2020년 3~5월 {}'.format(name),size=20)
    plt.legend(fontsize=13)
    plt.yticks(size=13)
    plt.show()
```

```python
# 미국 금리 그래프
draw_graph(amDF_total,'미국 금리')
```

###  중간 결론
   - 보통 금리는 시장이 안정됐을 때 인상되고, 불안정할 때 인하되는 경향이 있음.    
   - 2020년 3월쯤부터 코로나 대유행이 시작됐고, 연방준비제도(이하 연준)에서 금리 인하 결정을 내린 것은 2008년 금융위기 이후 처음.
   - 이후 평준화가 지속되다 FOMC(이하 연방공개시장위원회) 정례회의를 거쳐 2022년 3월과 5월 금리 인상 의견이 나와 반영이 됨.  
   - 현재는 시장이 적절히 반응하고 있고, 물가 안정의 복원은 꼭 필요한 일이라고 하면서 여러번의 금리 인상 가능성을 내비친 것으로 보아 계속 상승 곡선을 띌 것 같음.

## 2. KOSPI 데이터 2019~2022 전처리

```python
kospiDF=pd.read_csv('KOSPI2019~2022.csv',encoding='euc_kr')
kospiDF.head()
```

```python
# 결측치 있는지 확인
kospiDF.isnull().sum()
```

```python
dateDF1=kospiDF['일자'].apply(lambda x : x.split('/')[:3]) # 연, 월, 일 분리 ('/' 기준)
dateDF1=pd.DataFrame(dateDF1.tolist(),columns=('연','월','일')) # list 형태로 바꾼 걸 다시 DF로 생성
```

```python
# 연, 월, 일 분리한 DF에 원데이터 '종가' 컬럼 연결
kospiDF=pd.concat([dateDF1,kospiDF['종가']],axis=1)
kospiDF.head()
```

```python
# 2019년 데이터
kospiDF_2019=kospiDF[kospiDF.연=='2019']
# 2020년 데이터
kospiDF_2020=kospiDF[kospiDF.연=='2020']
# 2021년 데이터
kospiDF_2021=kospiDF[kospiDF.연=='2021']
# 2022년 데이터
kospiDF_2022=kospiDF[kospiDF.연=='2022']
kospiDF_2019.head()
```

```python
# 년도-월마다 평균
kospiDF_2019_mean=kospiDF_2019.groupby(by=['월']).mean()
kospiDF_2020_mean=kospiDF_2020.groupby(by=['월']).mean()
kospiDF_2021_mean=kospiDF_2021.groupby(by=['월']).mean()
kospiDF_2022_mean=kospiDF_2022.groupby(by=['월']).mean()
```

```python
# 코스피, 코스닥 둘 다 연도 붙여서 쓰려고 함수 만듦
def stock_year_mean(data,year,i=1):
    data['연']=year
    data=data.reset_index()
    if i==0: data['날짜']=data.apply(lambda x:str(x['연'])+'-'+x['월']+'-'+x['일'],axis=1) # 코스피 3-5월 데이터 따로
    else: data['날짜']=data.apply(lambda x:str(x['연'])+'-'+x['월'],axis=1)
    data=data.set_index('날짜')['종가']
    return data
```

```python
kospiDF_2019_mean=stock_year_mean(kospiDF_2019_mean,2019)
kospiDF_2020_mean=stock_year_mean(kospiDF_2020_mean,2020)
kospiDF_2021_mean=stock_year_mean(kospiDF_2021_mean,2021)
kospiDF_2022_mean=stock_year_mean(kospiDF_2022_mean,2022)
```

```python
kospiDF_total_mean=pd.concat([kospiDF_2019_mean,kospiDF_2020_mean,kospiDF_2021_mean,kospiDF_2022_mean])
kospiDF_total_mean.head()
```

## 4. 코스피와 코스닥

#### 코스피(KOSPI)란? 
: 종합주가지수를 뜻하는 단어로 우리나라 대표적인 주식시장.    
  네이버, 카카오, 삼성전자 등 대부분의 기업이 포함되어 있고, 가격 변동성이 낮은 것이 특징.
 
#### 코스닥(KOSDAQ)이란? 
: 우리나라 제2의 주식시장이고, 중소기업, 벤처기업들이 포함.    
  코스피에 반해 변동성이 크기 때문에, 하이 리스크 하이 리턴을 보이는 것이 특징.

### (1) 코스피와 코스닥 추이 그래프

```python
# 두 DF 비교 그래프 함수
def compare_draw_graph(data1,data2,name1,name2):    
    if(platform.system()=='Windows'):
        plt.rc('font',family='Malgun Gothic')
    else:
        plt.rc('font',family='AppleGothic')       
    plt.rcParams['axes.unicode_minus']=False
    plt.figure(figsize=(20,10))
    plt.title('2019~2022년 {}와 {} 추이'.format(name1,name2),size=20)
    if name1=='미국 금리': 
        plt.plot(data1,label=name1,color='tomato',marker='o')
        plt.ylabel('금리',size=15)# 왼쪽 Y축 이름
    else:
        plt.plot(data1,label=name1,color='tomato',marker='o')
        plt.annotate('<= 2020년 3월 코로나 대유행',xy=(14.3,1800),rotation=30,fontsize=16)
        plt.ylabel('종가',size=15)# 왼쪽 Y축 이름
    plt.xlabel('기간',size=15)
    plt.legend(loc=2)
    plt.yticks(size=13)
    plt.xticks(data1.index,rotation=60,size=13)
    y_right=plt.twinx()
    y_right.plot(data2,label=name2,color='royalblue',marker='s')
    plt.ylabel('종가',size=15)# 오른쪽 Y축 이름
    y_right.legend(loc=1)
    plt.yticks(size=13)
    plt.show()
```

```python
compare_draw_graph(kospiDF_total_mean,kosdaqDF_total_mean,'코스피','코스닥')
```

###  중간 결론

   - 코스피와 코스닥 둘 다 비슷한 추세를 보이며, 2020년 3월 급하락한 원인은 코로나 대유행의 영향으로 사료됨

```python
# 코스피와 코스닥 간의 상관계수
corrDF=pd.concat([kospiDF_total_mean,kosdaqDF_total_mean],axis=1)
corrDF.columns=['코스피','코스닥']
```

```python
corr=corrDF.corr()
corr
```

###  중간 결론
   - 코스피와 코스닥의 상관계수 또한 0.96으로 높게 나타나, 강한 양의상관관계가 있음. (그래프 흐름도 비슷한 경향을 나타냄)

### 
   - 코스피의 2020년 3월 하락은 2008년 금융위기 이후 처음. 이를 완화하고자 3월 19일 한-미 통화스와프 체결.
   - <b>통화스와프</b>란? 두 국가가 현재 환율에 따라 필요한 만큼의 돈을 교환하고, 일정 기간이 지난 후 최초 계약 때 정한 환율로 원금을 재교환하는 거래. 환율을 안정시키고, 경제 위축을 완화하는 역할. 사진을 보면 2020년 3월 이후부터 약 1년간 점점 환율이 내려간 것을 볼 수 있음.   
       (2008년 경제 위기 때와 2020년 코로나 대유행 때 실시)
   
   - 보통 환율이 오르면 원화의 가치가 낮아지고 달러의 가치가 오르기 때문에, 외국투자자들이 한국 투자금을 빼서 미국으로 가져감.     
     외국인 투자 비중이 높은 우리나라 증시에 악영향을 끼침 => 코스피 하락
     
   - 2020년 3월 이후는 반대인 상황이므로 그래프를 보면 코스피가 상승하고 있는 것을 볼 수 있음.

## 4. 미국 금리와 코스피, 코스닥 간의 상관관계
   - 보편적으로 금리가 하락하면 시중의 자금은 주식시장으로 몰리며 지수가 상승하는 반비례 경향 나타낸다고 함.
       - 이유: 금리가 낮아지면 기업들은 저금리로 자금을 빌려 투자를 늘리고, 동시에 부채를 안고 있는 기업들의 금융비용도 줄어들게 됨.    
          그럼 기업 이익은 늘고 재무구조 등이 튼튼해져 기업 가치가 오름. 그 결과로 주식을 사려는 수요가 늘고, 주가가 상승하기 때문.

```python
compare_draw_graph(amDF_total,kospiDF_total_mean,'미국 금리','코스피')
```

```python
compare_draw_graph(amDF_total,kosdaqDF_total_mean,'미국','코스닥')
```

```python
corrDF=pd.concat([amDF_total,kospiDF_total_mean,kosdaqDF_total_mean],axis=1)
corrDF.columns=['미국 금리','코스피','코스닥']
```

```python
corr=corrDF.corr()
corr
```

###  
   - 금리와의 상관계수를 살펴보면, 코스피(-0.67)와 코스닥(-0.73) 모두 음의 상관관계를 나타냄.
   - [번외] 7월은 아직 20일까지밖에 나오지 않아 데이터에 포함시키지 않았지만 평균을 구해 봤을 때 2,350 정도의 값으로 바로 이전인 6월 평균값인 2,474보다 낮은 것을 알 수 있었음.

### <span style='color:Blue'> [결론]   </span>
- 결과를 토대로 미국 금리와 코스피/코스닥은 하나가 증가하면 하나가 감소하는 역관계의 성향을 띄고,  최근 증가하는 금리로 인해 코스피/코스닥 지수가 한동안 낮을 것으로 예측됨.
